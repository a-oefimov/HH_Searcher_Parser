```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(executable_path=r"C:\Chromedriver\chromedriver.exe")
```

```python
detailed_data = {"Town":[], "Experience_years":[],"Education":[]}
```

```python
def ParseWorkExperience(link, detailed_data):
    
    driver.get(link)
            
    experience = driver.find_element(By.XPATH, '//span[@class="resume-block__title-text resume-block__title-text_sub"]')

    town = driver.find_element(By.XPATH, '//div[@class="bloko-translate-guard"]')

    education = driver.find_element(By.XPATH, '//div[@data-qa="resume-block-education"]')

    detailed_data["Town"].append(town.text)
    detailed_data["Experience_years"].append(experience.text)
    detailed_data["Education"].append(education.text)

    print(town.text)
    print(experience.text)
    print(education.text)
```
