```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
```

```python
URL_START = 'https://hh.ru/search/resume?text='

URL_MID = '&logic=normal&pos=workplaces&exp_period=all_time&exp_company_size=any&show_conditions=false&filter_exp_period=all_time&area='

URL_END = '&relocation=living_or_relocation&age_from=&age_to=&gender=unknown&salary_from=&salary_to=&currency_code=RUR&order_by=relevance&search_period=30&items_on_page=100&page='
```

```python
driver = webdriver.Chrome(executable_path=r"C:\Chromedriver\chromedriver.exe")
```

```python
search_region_info_HIRING = {"Волгоград": [24,7], "Ижевск": [96, 5], "Казань": [88,19], "Саратов": [79,8], "Тольятти": [212,4], "Самара": [78,14], "Ярославль": [112,6], "Нижний Новгород":[66,15]}
```

```python
def ParseHHJobs(query, search_regions_info, target_dict_for_df):
    preliminary_url = URL_START + query
    
    preliminary_url += URL_MID
    
    for region, region_data in search_regions_info.items():
        search_url = preliminary_url + str(region_data[0]) + URL_END
        
        for i in range(0, region_data[1]):
            driver.get(search_url + str(i))
            
            resumes = driver.find_elements_by_xpath('//div[@class="resume-card-content--NNaytYRNAe_OrXS4MB9e"]')

            print("Resumes", len(resumes))
            print("__________________________________")
            
            for elem in resumes:
                print(elem.text)
                print("Details:") 
                
                position_name = elem.find_element(By.XPATH, './/a[@class="magritte-link___b4rEM_4-3-2 magritte-link_style_accent___r-MW4_4-3-2 magritte-link_enable-visited___Biyib_4-3-2"]')
                target_dict_for_df["Desired_position_names"].append(position_name.text)
                print(position_name.text)

                target_dict_for_df["Links"].append(position_name.get_attribute('href'))
                print(position_name.get_attribute('href'))
                target_dict_for_df["Regions_list"].append(region)

                try:
                    desired_salary = elem.find_element(By.XPATH, './/div[@class="magritte-text___pbpft_3-0-15 magritte-text_style-primary___AQ7MW_3-0-15 magritte-text_typography-subtitle-2-semibold___3q3c-_3-0-15"]')
                    target_dict_for_df["Desired_salaries"].append(desired_salary.text)
                    print(desired_salary.text)
                except:
                    target_dict_for_df["Desired_salaries"].append("Не указана")
                    print("Не указана")

                try:
                    search_status = elem.find_element(By.XPATH, './/div[@class="magritte-tag__label___YHV-o_3-0-13"]')
                    target_dict_for_df["Search_Status"].append(search_status.text)
                    print(search_status.text)
                except:
                    target_dict_for_df["Search_Status"].append("Не указано")
                    print("Не указано")
                
                print("__________________________________")
```

```python
dict_for_df_HIRING = {"Desired_position_names":[], "Desired_salaries":[], "Search_Status":[], "Regions_list":[], "Links":[]}
ParseHHJobs("юрист", search_region_info_HIRING, dict_for_df_HIRING)
```
